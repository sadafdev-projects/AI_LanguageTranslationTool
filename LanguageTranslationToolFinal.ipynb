```python
from googletrans import Translator, LANGUAGES
from gtts import gTTS
import gradio as gr
import tempfile
```

```python
translator = Translator()
```

```python
def translate_text(text, src_lang, tgt_lang, tts_option):
    if not text.strip():
        return "âš ï¸ Please enter some text.", None

    try:
        # Perform translation
        translated = translator.translate(text, src=src_lang, dest=tgt_lang)
        translated_text = translated.text

        audio_path = None
        if tts_option:
            # Create a temporary audio file for playback
            tts = gTTS(translated_text, lang=tgt_lang)
            tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
            tts.save(tmp.name)
            audio_path = tmp.name

        return translated_text, audio_path

    except Exception as e:
        return f"âŒ Error: {str(e)}", None
```

```python
# Dropdown options for languages
language_options = list(LANGUAGES.items())
language_list = [(value.title(), key) for key, value in language_options]

# Interface
with gr.Blocks(title="AI Language Translator") as demo:
    gr.Markdown("## ğŸŒ AI Language Translation Tool")
    gr.Markdown("Enter text, choose source and target languages, and get instant translations!")

    with gr.Row():
        text_input = gr.Textbox(label="Enter text to translate", placeholder="Type something...")

    with gr.Row():
        src_lang = gr.Dropdown(choices=language_list, value="en", label="Source Language")
        tgt_lang = gr.Dropdown(choices=language_list, value="fr", label="Target Language")

    tts_option = gr.Checkbox(label="Enable Text-to-Speech", value=False)

    translate_btn = gr.Button("ğŸ” Translate")

    translated_output = gr.Textbox(label="Translated Text", placeholder="Your translation will appear here")
    audio_output = gr.Audio(label="Speech Output")

    translate_btn.click(
        translate_text,
        inputs=[text_input, src_lang, tgt_lang, tts_option],
        outputs=[translated_output, audio_output]
    )

demo.launch()
```
